```python
import numpy as np
import pandas as pd
from PIL import Image
```

```python
label_to_label_index_dict = {
    0: 1,
    1: 4,
    2: 8,
    3: 7,
    4: 6,
    5: 9,
    6: 3,
    7: 2,
    8: 5,
    9: 0
}

label_index_to_label_dict = {
    1:0,
    4:1,
    8:2,
    7:3,
    6:4,
    9:5,
    3:6,
    2:7,
    5:8,
    0:9
}
```

### X_20181226172954 & Y_20181226172954  - Error some label 1 should be 0

```python
X_file = 'X_20181226172954.npy'
Y_file = 'Y_20181226172954.npy'

X = np.load(X_file)
Y = np.load(Y_file)
```

```python
# Convert One-hot Label Index to Label
Y = np.argmax(Y, axis=1)
for i in range(len(Y)):
    Y[i] = label_index_to_label_dict[Y[i]]
```

```python
pd.DataFrame({'label':Y}).groupby('label')['label'].count()
```

```python
# Check Label
for i in range(len(Y)):
    if Y[i] == 1:
        display(Image.fromarray(X[i]))
```

```python
# Fix Label
invalid_label = 1
correct_label = 0
base_idx = 0
num_invalid = 15

for i in range(len(Y)):
    if Y[i] == invalid_label:
        if base_idx != 0:
            base_idx -= 1
        else:
            Y[i] = correct_label
            num_invalid -= 1
            
    if num_invalid == 0:
        break
```

```python
# Recheck Label
for i in range(len(Y)):
    if Y[i] == 1:
        display(Image.fromarray(X[i]))
```

```python
pd.DataFrame({'label':Y}).groupby('label')['label'].count()
```

```python
# Convert Label to Index
for i in range(len(Y)):
    Y[i] = label_to_label_index_dict[Y[i]]
```

```python
# Save Label
onehot_Y_buff = np.zeros((len(Y), 10))
onehot_Y_buff[np.arange(len(Y)), Y] = 1

np.save(open('crc_' + X_file,'wb'), np.array(X))
np.save(open('crc_' + Y_file,'wb'), onehot_Y_buff)
```
